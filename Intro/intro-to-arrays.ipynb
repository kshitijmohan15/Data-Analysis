How to find the unique fields in a data set using np.unique

```python
names = np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe'])
np.unique(names)
```

np.in1d, tests membership of the values in one array in another, returning a boolean array, as in it tells whether the given values are present in the array or not. intersect1d can be used to find out the common values, 

```python
np.in1d(names, ["Kshitij", "Bob", "Will"])
```
